# MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
x= np.arange(1,10);
y=4*x+2
plt.title('Matplotlib demo')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
## Computing x y coordinates using sine curve
x=np.arange(0,4* np.pi,0.1)
y=np.sin(x);
plt.title('Sine Wave ')

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,np.pi * 4,0.5)
y=np.cos(x);
plt.title('Cos Wave ')

plt.plot(x,y)
plt.show()
```

```python
## Subplot

x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title('Sin')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.show()
```

```python
## plot bargraph
x=[2,8,10,12]
y=[15,27,34,12]

x2=[3,9,11]
y2=[6,5,7]
plt.bar(x,y,color='r')
plt.bar(x2,y2,color='g')
plt.title('Bargraph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
## Histograms
a=np.array([22,87,5,56,77,1,44,12,6,10])
plt.hist(a,bins=7)
plt.title('Histograms')
plt.show()
```
